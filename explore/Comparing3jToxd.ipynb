```python
import numpy as np
from matplotlib import pyplot as plt
from quickRecoClass import quickReco
```

```python
#Loading DM and 3j files
DM = quickReco ('DM','\\Users\\student\\Desktop\\College\\Summer_2020\\Research\\DM_first_model_first_run\\ppTOllxdxd_signal')
PJ = quickReco('PJ','\\Users\\student\\Desktop\\College\\Summer_2020\\Research\\DM_first_model_first_run\\ppTO3j background')
```

```python
PJ_sht_array=[]

for i,arr in enumerate (PJ.l_scalarHT):
    for j,value in enumerate(arr):
        PJ_sht_array.append(value)
```

```python
DM_sht_array=[]

for i,arr in enumerate (DM.l_scalarHT):
    for j,value in enumerate(arr):
        DM_sht_array.append(value)
```

```python
plt.hist(PJ_sht_array, bins=np.arange(0,1000,10), alpha=0.4, label='SM ll+jjj')
plt.hist(DM_sht_array, bins=np.arange(0,1000,10), alpha=0.4, label='2HDMa ll+xdxd')
plt.title('Scalar HT')
plt.xlabel('Scalar HT [GeV]')
plt.ylabel('Entries / Bin')
plt.legend()
plt.show()
```
